```python
import pandas as pd
import numpy as np
import h5py

# Load CSV
df = pd.read_csv("merge.csv")

# Filter earthquakes with valid p_arrival_samples and magnitude > 1
df_eq = df[(df.trace_category == 'earthquake_local') & 
           (df.source_magnitude > 1) & 
           (~df.p_arrival_sample.isna()) &  # Not NaN
           (df.p_arrival_sample > 0)        # Greater than 0
          ].sample(n=15000, random_state=42, replace=False)

# Sample noise data
df_noise = df[df.trace_category == 'noise'].sample(n=10000, random_state=42, replace=False)

# Combine the dataframes
combined_df = pd.concat([df_eq, df_noise], ignore_index=True)

# Prepare to load waveforms
H5_PATH = "merge.hdf5"
all_trace_names = combined_df["trace_name"].tolist()
num_waveforms = len(all_trace_names)

# Create empty arrays to store waveforms and their metadata
waveforms = []
valid_trace_names = []

# Load waveforms while keeping track of which ones were successfully loaded
with h5py.File(H5_PATH, "r") as h5_file:
    for trace_name in all_trace_names:
        hdf5_path = f"data/{trace_name}"
        if hdf5_path in h5_file:
            waveform = np.array(h5_file[hdf5_path])
            waveforms.append(waveform)
            valid_trace_names.append(trace_name)
        else:
            print(f"Warning: {hdf5_path} not found in HDF5 file.")

# Convert to numpy array
X_numpy_waveform = np.array(waveforms, dtype=np.float32)

# Filter the dataframe to only include traces that were successfully loaded
filtered_df = combined_df[combined_df["trace_name"].isin(valid_trace_names)]

# Now we can shuffle while maintaining the connection between metadata and waveforms
shuffle_indices = np.random.permutation(len(X_numpy_waveform))
X_numpy_waveform_shuffled = X_numpy_waveform[shuffle_indices]
filtered_df_shuffled = filtered_df.iloc[shuffle_indices].reset_index(drop=True)

print(f"Loaded {len(X_numpy_waveform_shuffled)} total waveforms with shape: {X_numpy_waveform_shuffled.shape}")
print(f"Number of earthquakes: {len(filtered_df_shuffled[filtered_df_shuffled.trace_category == 'earthquake_local'])}")
print(f"Number of noise samples: {len(filtered_df_shuffled[filtered_df_shuffled.trace_category == 'noise'])}")
print(f"Number of NaN p_arrivals: {filtered_df_shuffled['p_arrival_sample'].isna().sum()}")
print(f"Number of non-zero, non-NaN p_arrivals: {filtered_df_shuffled['p_arrival_sample'].dropna().ne(0).sum()}")
```

```python
import os
import numpy as np
import pandas as pd

SAVE_DIR = "seismic_dataset"
waveform_dir = os.path.join(SAVE_DIR, "waveforms")

# Check if the waveform directory exists
if not os.path.exists(waveform_dir):
    raise FileNotFoundError(f"❌ Waveform directory not found: {waveform_dir}")

metadata_path = os.path.join(SAVE_DIR, "metadata.csv")

if not os.path.exists(metadata_path):
    raise FileNotFoundError(f"❌ Metadata CSV not found: {metadata_path}")

metadata_df = pd.read_csv(metadata_path)

print(f"✅ Loaded metadata with {len(metadata_df)} records.")
print(metadata_df.head())

# Load all waveforms
waveforms = []

for file_name in metadata_df["file_name"]:
    file_path = os.path.join(waveform_dir, file_name)
    
    if os.path.exists(file_path):
        waveform = np.load(file_path)  # Load the .npy file
        waveforms.append(waveform)
    else:
        print(f"⚠️ Warning: File not found {file_name}")

waveforms = np.array(waveforms, dtype=object)  

print(f"✅ Loaded {len(waveforms)} waveforms.")
print(f"Example waveform shape: {waveforms[0].shape}")  # Display shape of first waveform

trace_name = metadata_df["file_name"].iloc[0].replace(".npy", "")  # Take the first trace as an example
file_name = f"{trace_name}.npy"

if file_name in metadata_df["file_name"].values:
    file_path = os.path.join(waveform_dir, file_name)
    
    if os.path.exists(file_path):
        waveform = np.load(file_path)
        print(f"✅ Loaded waveform: {trace_name}, Shape: {waveform.shape}")
    else:
        print(f"❌ File {file_name} not found.")
else:
    print(f"❌ Trace name {trace_name} not found in metadata.")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import Sequence
```

```python
class SeismicDataGenerator(Sequence):
    def __init__(self, X_waveforms, y_waveforms, window_size, batch_size=32, shuffle=True, norm_method="z-score"):
        self.X_waveforms = self.normalize_waveform(X_waveforms)  # Apply normalization first
        self.y_waveforms = y_waveforms  # Full labels
        self.window_size = window_size  # Window size for segmentation
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.indices = np.arange(len(self.X_waveforms))
        
        if self.shuffle:
            np.random.shuffle(self.indices)
    
    def __len__(self):
        return int(np.floor(len(self.X_waveforms) / self.batch_size))  # Number of batches per epoch
    
    def __getitem__(self, index):
        batch_indices = self.indices[index * self.batch_size:(index + 1) * self.batch_size]
        
        X_batch = [self.extract_window(self.X_waveforms[i]) for i in batch_indices]
        y_batch = [self.extract_window(self.y_waveforms[i]) for i in batch_indices]
        
        X_batch = np.array(X_batch, dtype=np.float32)
        y_batch = np.array(y_batch, dtype=np.float32)
        
        return X_batch, y_batch
    
    def extract_window(self, waveform):
        start_idx = np.random.randint(0, len(waveform) - self.window_size + 1)
        return waveform[start_idx:start_idx + self.window_size]
    
    def normalize_waveform(self, waveforms):
        if self.norm_method == "z-score":
            mean = np.mean(waveforms, axis=(1, 2), keepdims=True)
            std = np.std(waveforms, axis=(1, 2), keepdims=True) + 1e-8  # Avoid division by zero
            return (waveforms - mean) / std
        elif self.norm_method == "min-max":
            min_val = np.min(waveforms, axis=(1, 2), keepdims=True)
            max_val = np.max(waveforms, axis=(1, 2), keepdims=True)
            return (waveforms - min_val) / (max_val - min_val + 1e-8)
        else:
            return waveforms  # No normalization
    
    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.indices)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def self_attention_block(x, filters):
    """ Self-attention mechanism for feature refinement. """
    f = layers.Conv1D(filters, 1, padding="same")(x)  # Key
    g = layers.Conv1D(filters, 1, padding="same")(x)  # Query
    h = layers.Conv1D(filters, 1, padding="same")(x)  # Value
    
    attention = layers.Activation("softmax")(tf.matmul(g, f, transpose_b=True))
    attention_output = tf.matmul(attention, h)
    
    return layers.Add()([x, attention_output])

def conv_block(x, filters, strides=1):
    """ Convolutional block with stride for downsampling. """
    x = layers.Conv1D(filters, 3, strides=strides, padding="same")(x)
    x = layers.BatchNormalization()(x)
    return layers.Activation("relu")(x)

def upsample_block(x, filters):
    """ Transposed convolution for upsampling. """
    x = layers.Conv1DTranspose(filters, 3, strides=2, padding="same")(x)
    x = layers.BatchNormalization()(x)
    return layers.Activation("relu")(x)

def create_unetpp_model(input_shape, filters=64):
    inputs = layers.Input(input_shape)
    
    # Add channel dimension if necessary
    if len(input_shape) == 2:
        x = inputs
    else:
        x = layers.Reshape(input_shape + (1,))(inputs)
    
    # Encoder with strided convolutions
    c1 = conv_block(x, filters)
    c2 = conv_block(c1, filters * 2, strides=2)
    c3 = conv_block(c2, filters * 4, strides=2)
    c4 = conv_block(c3, filters * 8, strides=2)
    
    # Self-attention bottleneck
    c5 = self_attention_block(conv_block(c4, filters * 16, strides=2), filters * 16)
    
    # Decoder path with self-attention
    u4 = upsample_block(c5, filters * 8)
    c4 = self_attention_block(c4, filters * 8)
    m4 = layers.Concatenate()([u4, c4])

    u3 = upsample_block(m4, filters * 4)
    c3 = self_attention_block(c3, filters * 4)
    m3 = layers.Concatenate()([u3, c3])

    u2 = upsample_block(m3, filters * 2)
    c2 = self_attention_block(c2, filters * 2)
    m2 = layers.Concatenate()([u2, c2])

    u1 = upsample_block(m2, filters)
    c1 = self_attention_block(c1, filters)
    m1 = layers.Concatenate()([u1, c1])

    # Output layer
    outputs = layers.Conv1D(1, 1, activation="sigmoid")(m1)
    outputs = layers.Reshape((input_shape[0],))(outputs)

    # Create model
    model = models.Model(inputs=inputs, outputs=outputs)

    # Compile model
    model.compile(
        optimizer="adam",
        loss="binary_crossentropy",
        metrics=["mae", tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
    )

    return model

# Example usage:
model = create_unetpp_model(input_shape=(128, 3))
model.summary()
```
